```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
colors = plt.cm.tab10.colors
import os
```

```python
path = '../Examples/Gas/'
```

```python
conv = np.loadtxt(path + 'static/convergence')
fig,axs = plt.subplots(5,1, sharex=True)
fig.subplots_adjust(hspace=0)
labels = ['energy','energy_diff','abs_dens','rel_dens','abs_ev','rel_ev']
for i in range(5):
    axs[i].plot(conv[:,0], conv[:,i+1], 'o-', label = labels[i], color = colors[i])
    axs[i].set_xlabel('Number of iterations')
    axs[i].set_yscale('log')
fig.legend(ncol = 5)
```

```python
dipole_acc = np.loadtxt(path+'/td.general/acceleration')
energy = np.loadtxt(path+'/td.general/energy')
laser = np.loadtxt(path+'/td.general/laser')
multipoles = np.loadtxt(path+'/td.general/multipoles')

fig,axs = plt.subplots(5,1, sharex=True)
fig.subplots_adjust(hspace=0.1)
labels = ['dipole_acc','energy','laser','multipoles']
axs[0].plot(dipole_acc[:,1], dipole_acc[:,2], '-', label = labels[0], color = colors[0])
axs[1].plot(energy[:,1], energy[:,2], '-', label = labels[1], color = colors[1])
axs[2].plot(laser[:,1], laser[:,2], '-', label = labels[2], color = colors[2])
axs[3].plot(multipoles[:,1], 1 - multipoles[:,2], '-', label = '1 - N$_{e}$', color = colors[3])
axs[4].plot(multipoles[:,1], multipoles[:,3], '-', label = labels[3], color = colors[4])
fig.legend(ncol = 5, loc = 'upper center')    
```

```python
hhg = np.loadtxt(path+'/hs-mult.x')
plt.plot(hhg[:,0]/0.05655, np.log10(hhg[:,1]), label = 'HHG', color = 'k', linewidth = 2)
for x in range(1, int(hhg[:,0].max()/0.05655), 2,):
    plt.axvline(x=x, color='gray', linestyle='--', linewidth=0.5)
plt.xlabel('Harmonic Order')
plt.ylabel('log$_{10}$(Intensity)')
plt.legend()
```
