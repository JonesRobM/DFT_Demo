```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib import cm
import glob
colors = plt.cm.tab10.colors
import os
import sys
path = '../Examples/TransientAbsorption/'
```

```python
laser = np.loadtxt(path+'td.general_ref/laser')
plt.plot(laser[:,1],laser[:,2])
plt.text(0.125,0.85,'A(t)',color='k',transform=plt.gca().transAxes, fontsize=16)
```

```python
ta = np.loadtxt(path+'result.dat')
x = sorted(set(ta[:,0]))
y = sorted(set(ta[:,1]))
Map = np.zeros((len(x), len(y)))

for i, item in enumerate(ta):
    Map[x.index(item[0]), y.index(item[1])] = item[2]

fig, ax = plt.subplots()
im = ax.imshow(Map.T, cmap='seismic', 
           extent=[min(x), max(x), max(y), min(y)], 
           aspect='auto')
cbar_ax = fig.add_axes([0.95, 0.11, 0.025, 0.77])
cax = fig.colorbar(im, cax=cbar_ax)
ax.set_ylabel('Energy (a.u)')
ax.set_xlabel('Time (a.u.)')
ax2 = ax.twinx()
ax2.plot(laser[:,1],1-0.25*laser[:,2],color='k')
ax2.set_xlim([min(x), max(x)])
ax2.set_ylim([-1,10])
ax2.text(0.125,0.125,'A(t)',color='k',transform=plt.gca().transAxes, fontsize=16)
```
